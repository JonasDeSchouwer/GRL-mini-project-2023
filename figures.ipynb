```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
acc_1a = np.array([0.6645,0.6571,0.6253,0.5641,0.4318,0.382,0.3404,0.2914,0.2082,0.1837])
acc_1b = np.array([1,1,1,1,1,0.8441,0.4465,0.258,0.1543,0.2073])
acc_2a = np.array([0.7592,0.7184,0.6996,0.6416,0.6384,0.5086,0.4098,0.3543,0.2743,0.2171])
acc_2b = np.array([1,1,1,1,1,0.8376,0.5273,0.4212,0.3004,0.2131])
acc_argmax = np.array([1,1,1,1,1,0.8449,0.441,0.2571,0.1396,0.06])
```

```python
fig = plt.figure()
plt.plot(acc_argmax, '-^k', label=r"argmax", zorder=10)
plt.plot(acc_1a, '--.', label=r"1-layer $\mathcal{A}$")
plt.plot(acc_2a, '--.', label=r"2-layer $\mathcal{A}$")
plt.plot(acc_1b, '--.', label=r"1-layer $\mathcal{B}$")
plt.plot(acc_2b, '--.', label=r"2-layer $\mathcal{B}$")
plt.xlabel(r"$\angle (\boldsymbol{x}_i, \boldsymbol{t}_i)$")
plt.ylabel("test accuracy")
plt.xticks(ticks=list(range(10)), labels=[str(10*s)+'°' for s in range(10)])
plt.legend()
```
